```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python
# dados CSV
df = pd.read_csv('/content/Arquivos Base/bike-sharing.csv')
df = df.set_index('datetime') # seta uma coluna como index
df.sort_index() # ordena por indice
df.head(3)
```

```python
# Tamanho Dataframe
df.shape
```

```python
# Media coluna windspeed
df.windspeed.mean()
```

```python
# Media coluna Temp
df.temp.mean()
```

```python
# quantos registro existem para o ano de 2011
df[(df.index >= '2011-01-01') & (df.index < '2012-01-01')]
```

```python
# quantos registro existem para o ano de 2012
df[(df.index >= '2012-01-01') & (df.index < '2013-01-01')]
```

```python
# quantos bicicletas foram alugadas em 2011
x = df[(df.index >= '2011-01-01') & (df.index < '2012-01-01')] 
x.total_count.sum()
```

```python
# quantos bicicletas foram alugadas em 2012
x = df[(df.index >= '2012-01-01') & (df.index < '2013-01-01')] 
x.total_count.sum()
```

```python
# media de locações por temporada
df.groupby(['season'])["total_count"].mean()
```

```python
# media de locações por hora
x = df.groupby(['hour'])["total_count"].mean()
x.sort_values()
```

```python
# media de locações por semana
x = df.groupby(['weekday'])["total_count"].mean()
x.sort_values()
```

```python
# media de locações as quartas
quarta = df.groupby(['weekday']).get_group(3)
# display(x)
x = quarta.groupby(['hour'])["total_count"].mean()
x.sort_values()
```

```python
# media de locações as quartas
quarta = df.groupby(['weekday']).get_group(6)
# display(x)
x = quarta.groupby(['hour'])["total_count"].mean()
x.sort_values()
```
