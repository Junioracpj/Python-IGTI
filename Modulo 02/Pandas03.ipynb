```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python
# dados CSV
df = pd.read_csv('https://pycourse.s3.amazonaws.com/temperature.csv')
df = df.set_index('date') # seta uma coluna como index
# df.sort_index()
display(df)
```

```python
# GroupBY agrupamento por valores unicos de uma ou mais colunas
df.groupby(by='classification')
df.groupby(by='classification').mean()
```

```python
df.groupby(by='classification').sum()
# exibi a soma de todos os grupos
```

```python
# Drop - Remocao de uma coluna ou linha
df.drop('temperatura', axis=1)
```

```python
df2 = df.copy() # cria uma copia independente
df2.drop('temperatura', axis=1)
df2.drop('temperatura', axis=1, inplace=True) #substitui o original
df2.head()
```
