```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python
# dados CSV
df = pd.read_csv('https://pycourse.s3.amazonaws.com/temperature.csv')
df = df.set_index('date') # seta uma coluna como index
df.sort_index()
display(df)
```

```python
#plot de linhas
# index e considerado o eixo do plot
df.plot(); 
```

```python
df.plot(figsize=(5, 5)); #horizontal e vertical
```

```python
df.plot(figsize=(5,5), grid=True);
#linhas grades no grafico
```

```python
# df.plot(style='-o', figsize=(5,5), grid=True);
df.plot(style='--', figsize=(5,5), grid=True);
# df.plot(style='-', figsize=(5,5), grid=True);
```

```python
df.plot(style='--', linewidth=5, figsize=(5,5), grid=True);
```

```python
df.plot(style='--', linewidth=5, color='red', figsize=(5,5), grid=True);
```

```python
#plot de barras
df['classification'].value_counts().plot.bar(figsize=(10,5),rot=0);
```

```python
#plot de barras
df.plot(kind='bar', figsize=(10,5), rot=30)
```

```python
df['classification'].value_counts().plot.pie(autopct='%1.1f%%',
                                             shadow=True,
                                             figsize=(10,7));
```
