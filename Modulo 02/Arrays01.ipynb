Slicing retorna o fateamento de um array, sempre indicando o inicio e o fina do fatiamento

```python
num[1:4] # ira retonar os valores entre 1 e 4, excluindo o ultimo
b[1:3, 1:4] # em nivel 2D, sera a mesma coisa
```

```python
x = np.linspace(start=10, stop=100, num=10) # cra um array com tamanho e intercalacao definida
print('x:', x)
print('shape: ', x.shape) # retorna o tamanho do array
```

Extracao de elementos em Arrays

```python
print(x[0])
print(x[1])
print(x[4])
print(x[-1])
print(x[-3]) #do final para o inicio
```

```python
print(x[0:3]) # 3 primeiros elementos
print(x[:3]) # 3 primeiros elementos tbm
print(x[-2:]) # 2 ultimos primeiros numeros
```

```python
x = x.reshape(2, 5) # redimensiona o array para L e C (matriz) de acordo com seu tamanho
print(x)
print(x[0, 1])
print(x[1, -2])
print(x[1, 4])
```

```python
print(x[0:1, 1:3]) # slicing em arrays com duas dimensoes
print(x[0, 1:]) # tbm fatiamento, porem com definicoes diferentes
```

Compartilhamento de memoria do array

```python
x = np.array([1, 2, 3])
y = x[:2] # gerara na memoria, o mesmo valor para x e y, ficando uma shadow copy
y[0] = -100 #a mudanca sera efetuada para ambas as variaveis
z = y.copy() # com esse metodo, as variaveis ficaram independentes
z = 2
print(x, y, z)
```
