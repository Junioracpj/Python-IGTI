Operacoes em Arrays - Sempre vai executar a operacao em elemento a elemento, com indices equivalentes as duas variaveis

```python
import numpy as np
np.add # operacao de soma
np.subtract #subtracao
np.divide # divisao
np.multiply # multiplicacao
```

Broadcasting - operacoes onde nao existem arrays equivalentes, fazendo uma operacao mais direta
Economiza memoria, pois e mais direto em sua funcao

Operacoes matriciais

```python
x = a @ b #python puro 
np.dot(a, b) #numpy
a.dot(b) #numpy
# todas se referem a multiplicacao de matrizes
# nesse caso, sera uma geracao de novas matrizes a partir de uma antiga
```

```python
x = np.ones((2, 2)) # cria arrays preenchidos com 1
y = np.eye(2) # cria arrays com preenchidos com 1 na diagonal principal
print(x)
print(y)
```

```python
#soma\subtracao\divisao
print(x + y) # feito a elemento a elemento
print(x + 2) # o segundo elemento vai somar com todos do primeiro BROADCASTING
```

```python
# quando o broadcasting nao funciona
np.array([1, 2, 3,]) + np.array([1, 2])
# o erro ocorre por conta da inconsistencia de numeros de colunas ou linhas
```

```python
# multiplicacao
print(x * y) # feito a elemento a elemento
print(x * 2) # o segundo elemento vai dividir com todos do primeiro BROADCASTING
5 * np.ones([10,10]) # cria arrays com ones mas de diversos valores
```

```python
# multiplicacao matricial
print(np.dot(x, y))
print(x @ y)
print(x.dot(y)) # todas executam a mesma operacao
```

Comparacoes em Arrays

```python
a = np.array([1, 2, 3])
b = np.array([3, 4, 2])
s = 3

print( a < b) # vai retornar valores boleanos de cada comparacao de indices equivalentes
print( a > s) # vai comparar a segunda variavel com todos os elementos do array
print( a <= b) # mesma coisa, mas com duas condicionantes

print( a == b) # comparacao de igualdade em arrays
print( a == s) # comparacao de igualdade em broadcasting
```
