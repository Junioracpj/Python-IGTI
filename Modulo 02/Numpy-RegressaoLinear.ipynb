```python
import matplotlib.pyplot as plt # vizualicao de dados
import numpy as np
```

```python
# dados
x = [-1, -0.77777778, -0.55555556, -0.33333333, -0.11111111,
     0.1111111111, 0.33333333, 0.55555556, 0.77777778, 1.]
y = [ -1.13956201, -0.5717999, -0.21697033, 0.5425699, 0.49406657,
     1.14972239, 1.64228553, 2.1749824, 2.64773614, 2.9568202]
```

```python
#plot de dados
plt.figure(figsize=(10, 5)) # define o tamanho
plt.plot(x, y, 'o', label='dados originais') # tipo de plot
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
# transformando para numpy e vetor coluna
x, y = np.array(x).reshape(-1, 1), np.array(y).reshape(-1, 1)

# adcionando bias =  para estimar o termo b
z = np.hstack((x, np.ones(x.shape))) # cria o termo b e junta em um array

# estimando a e b
beta = np.linalg.pinv(z).dot(y)
print('a estimado:', beta[0][0])
print('b estimado:', beta[1][0])
# print(z)
```

```python
plt.figure(figsize=(10, 5)) # define o tamanho
plt.plot(x, y, 'o', label='dados originais')
plt.plot(x, z.dot(beta), label='regressao linear') # plotando a regressao linear
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regressao linear com numpy')
plt.grid()
plt.show()
```
