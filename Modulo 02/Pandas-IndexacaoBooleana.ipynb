Indexacao Booleana

```python
import numpy as np

a = np.array([1, 2, 3])
cond = a <= 2 # a condicao vai retornar valores boleanos em array
d = a[cond] # usando como parametro, o d vai conter somente os valores verdadeiros para a condicao

print(a)
print(cond)
print(d)
```

```python
x = np.array([[1, 2], [3, 4]])
y = np.array([1.5, 3.5])
print(x, y)

print( x > 2)
print( x >= 2)
print( x > x)
print( x > y) # comparacao entre arrays diferentes, BROADCASTING
```

```python
z = np.array([[1, 3,7],
              [4, 11, 21],
              [42, 8, 9]])
print(z)
```

```python
# indexacao booleana, retorna os numero de elementos maiores que K
k = 8
cond = z > k
print(cond)
print(f'elementos maiores que {k}:', z[cond])
print(f'numero de  elementos maiores que {k}:', len(z[cond]))
```

```python
cond = z % 2 == 0 # operacoes mais complexas de condicao
print(cond)
print(z[cond])
```
